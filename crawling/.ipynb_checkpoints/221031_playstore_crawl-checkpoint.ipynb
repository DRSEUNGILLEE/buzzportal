```python
def scroll(modal):
    try:        
        # 스크롤 높이 받아오기
        last_height = driver.execute_script("return arguments[0].scrollHeight", modal)
        for i in range(10):
            pause_time = random.uniform(0.5, 0.8)
            # 최하단까지 스크롤
            driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight);", modal)
            # 페이지 로딩 대기
            time.sleep(pause_time)
            # 무한 스크롤 동작을 위해 살짝 위로 스크롤
            driver.execute_script("arguments[0].scrollTo(0, arguments[0].scrollHeight-50);", modal)
            time.sleep(pause_time)
            # 스크롤 높이 새롭게 받아오기
            new_height = driver.execute_script("return arguments[0].scrollHeight", modal)
            try:
                # '더보기' 버튼 있을 경우 클릭
                all_review_button = driver.find_element_by_xpath('/html/body/div[1]/div[4]/c-wiz/div/div[2]/div/div/main/div/div[1]/div[2]/div[2]/div/span/span').click()
            except:
                # 스크롤 완료 경우
                if new_height == last_height:
                    print("스크롤 완료")
                    break
                last_height = new_height
                
    except Exception as e:
        print("에러 발생: ", e)
        

        
# 크롬 드라이버 세팅

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument("--single-process")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--log-level=3")
chrome_options.add_argument('--disable-logging')
chrome_options.add_argument('–-lang=ko')
chrome_options.add_argument("accept-language=ko")
driver = webdriver.Chrome('/home/buzzportal/crawling/chromedriver',options=chrome_options)

# 페이지 열기
URL = "https://play.google.com/store/apps/details?id=com.miraeasset.trade&hl=ko"
driver.get(URL)

time.sleep(3)

html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")   
time.sleep(3)


acont = driver.find_element(By.CSS_SELECTOR,'div.qZmL0')
ac = acont.find_elements(By.XPATH, "./child::*")
rp = ac[3]
review = rp.find_element(By.CSS_SELECTOR,'button.VfPpkd-Bz112c-LgbsSe.yHy1rc.eT1oJ.QDwDD.mN1ivc.VxpoF')

time.sleep(10)

driver.execute_script("arguments[0].click();", review)

time.sleep(5)


acont = driver.find_element(By.CSS_SELECTOR,'div.jgIq1')
ac = acont.find_elements(By.XPATH, "./child::*")
rp = ac[1]
scroll(rp)

html = driver.page_source

tree = etree.parse(StringIO(html),parser = etree.HTMLParser()) 

for i in range(1, 101):
    item = tree.xpath("/html/body/div[4]/div[2]/div/div/div/div/div[2]/div/div[1]/div[{}]/div[1]".format(str(i)))[0].text
    user = tree.xpath('/html/body/div[4]/div[2]/div/div/div/div/div[2]/div/div[1]/div[{}]/header/div[1]/div[1]/div'.format(str(i)))[0].text
    date = tree.xpath('/html/body/div[4]/div[2]/div/div/div/div/div[2]/div/div[1]/div[{}]/header/div[2]/span'.format(str(i)))[0].text
    
    dt = datetime.datetime.strptime(date.strip(), "%Y년 %m월 %d일")
    star = tree.xpath("/html/body/div[4]/div[2]/div/div/div/div/div[2]/div/div[1]/div[{}]/header/div[2]/div".format(str(i)))[0].attrib['aria-label']
    star = star[10]
    
    pos = get_sentiment(item)
    data = {'SOURCE': "Playstore_m.Stock", 'RELATED_ITEM': '미래에셋증권', 'DATETIME': dt.strftime("%y%m%d") + "-000000", 'TITLE': item ,'CONTENT': item, 'POSITIVITY': str(pos), 'NUM_REPLY': None, 'URL': None, 'CRAWLING_DATETIME': datetime.datetime.today().strftime("%Y%m%d-%H%M%S") , 'OTHERS' : {'STAR': star}}
    df= pd.DataFrame(data, index=[0])

    data_df = data_df.append(data, ignore_index=True)
#     df.to_csv('playstore_221031_1.tsv',sep="\t") 
    data_df['OTHERS'] = data_df['OTHERS'].apply(lambda x : str(x))
#     data_df.to_sql('TABLE_BUZZ',conn,if_exists = 'append',index = False)
    conn = sqlite3.connect("/home/buzzportal/crawling/opinion_outside.db")
    cur = conn.cursor()

    df_already= pd.read_sql("SELECT * FROM TABLE_BUZZ where SOURCE  = 'Playstore_m.Stock' ORDER BY DATETIME desc",conn)

    df_ttl = data_df
    df_ttl = df_ttl[df_ttl['DATETIME'] > df_already['DATETIME'].max()].reset_index(drop=True)

    if len(df_ttl) >0:
        df_ttl = df_ttl.reset_index(drop=True)
        df_ttl.to_sql('TABLE_BUZZ',conn,if_exists = 'append',index = False)
        print((datetime.datetime.now() + datetime.timedelta(hours=9)).strftime('%Y%m%d-%H%M%S')  , " m.Stock(PlayStore) // SUCCESS  : ",len(df_ttl))
    else:
        print((datetime.datetime.now() + datetime.timedelta(hours=9)).strftime('%Y%m%d-%H%M%S') , " m.Stock(PlayStore)//  NO DATA TO UPDATE")





        
```
